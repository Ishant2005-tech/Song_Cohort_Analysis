## ðŸ”¹ Identification of Popular Tracks Using Percentile Thresholding

A popularity threshold based on the 75th percentile was used to classify tracks as popular or non-popular.  
By aggregating popular tracks at the album level, this analysis highlights albums with a higher concentration of well-performing songs, providing insight into overall album impact and consistency.


```python
popularity_threshold = df['popularity'].quantile(0.75)
print(f"75th percentile of popularity: {popularity_threshold}")

df['is_popular'] = df['popularity'] > popularity_threshold

popular_songs_per_album = df.groupby('album')['is_popular'].sum().sort_values(ascending=False).reset_index()

print("\nNumber of popular songs per album (top 5):\n")
print(popular_songs_per_album.head())
```

## ðŸ”¹ Feature Engineering: Decade-Based Release Categorization

A new categorical feature was created by grouping songs into release decades based on their release year.  
This transformation enables decade-level analysis, allowing temporal patterns and stylistic shifts to be examined more effectively during exploratory analysis and clustering.


```python
def get_decade(year):
    if pd.isna(year):
        return 'Unknown'
    start_year = int(year / 10) * 10
    return f'{start_year}s'

df['decade'] = df['release_year'].apply(get_decade)
display(df.head())
```

## ðŸ”¹ Decade-Wise Correlation Analysis with Popularity

Correlations between key musical features and track popularity were computed separately for each release decade.  
This decade-wise analysis helps reveal how the influence of audio characteristics on popularity has evolved over time, providing temporal context that can inform feature selection and clustering strategies.


```python
numerical_features = [
    'acousticness', 'danceability', 'energy', 'instrumentalness',
    'liveness', 'loudness', 'speechiness', 'tempo', 'valence',
    'duration_ms'
]

decades = sorted(df['decade'].dropna().unique())

correlation_data = {}

for decade in decades:
    decade_df = df[df['decade'] == decade]
    if not decade_df.empty and len(decade_df) > 1: 
        
        decade_correlations = decade_df[numerical_features + ['popularity']].corr()['popularity']
        correlation_data[decade] = decade_correlations[numerical_features]


correlation_df = pd.DataFrame(correlation_data).T


correlation_df.index = pd.CategoricalIndex(correlation_df.index, categories=decades, ordered=True)
correlation_df = correlation_df.sort_index()

print("Correlations of musical features with popularity per decade:")
display(correlation_df.head())
```

## ðŸ”¹ Temporal Trends in Featureâ€“Popularity Relationships

Line plots were used to visualize how the correlation between musical features and track popularity changes across decades.  
This analysis highlights shifting listener preferences over time and reveals which audio characteristics have gained or lost influence on popularity, providing valuable context for clustering and similarity analysis.


```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 10))


features_to_plot = correlation_df.columns


n_cols = 3
n_rows = (len(features_to_plot) + n_cols - 1) // n_cols

for i, feature in enumerate(features_to_plot):
    plt.subplot(n_rows, n_cols, i + 1)
    sns.lineplot(x=correlation_df.index, y=correlation_df[feature], marker='o')
    plt.title(f'Popularity Correlation with {feature.replace("_", " ").title()}', fontsize=12)
    plt.xlabel('Decade', fontsize=10)
    plt.ylabel('Correlation with Popularity', fontsize=10)
    plt.xticks(rotation=45, ha='right')
    plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
